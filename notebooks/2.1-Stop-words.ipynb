```python
from collections import Counter

import pandas as pd
from src.data.rnn_preprocessing import clean_company_name_string
```

```python
data = pd.read_csv('../data/raw/train.csv')
```

```python
names = list(data['name_1'].values) + list(data['name_2'])
names = sorted(set(names))
```

```python
clean_names = [clean_company_name_string(n) for n in names]
print(f'Всего названий {len(clean_names)}')
```

```python
words = []
for n in clean_names:
    words.extend(n.split())
```

```python
stop_words = [word for word, cnt in Counter(words).most_common() if cnt > 70]
len(stop_words)
```
